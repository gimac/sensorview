```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.io import loadmat

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
sns.set_style('darkgrid')
```

```python
data_set = 'ARDP003R1_HW101_20190520_NewAF'
path = './data/'+data_set+'/'

files = []
for r, d, f in os.walk(path):
    for file in f:
        if '.pkl' in file:
            files.append(file)
    break
```

```python
det_list = pd.DataFrame()

for f in files:
    det_list = det_list.append(pd.read_pickle(path+f), ignore_index=True)

# det_list = pd.read_pickle(path+files[0])
len(det_list)
```

```python
ax = sns.barplot(x="AFType",
                 y="Frame",
                 #                  hue="LookName",
                 data=det_list,
                 estimator=lambda x: len(x) / len(det_list)*100)
ax.set(ylabel="Percentage")
plt.title(data_set)
plt.savefig(data_set+'_aftype.svg')
```

```python
ax = sns.barplot(x="AzConf",
                 y="Frame",
#                  hue="LookName",
                 data=det_list,
                 estimator=lambda x: len(x) / len(det_list)*100)
ax.set(ylabel="Percentage")
plt.title(data_set)
plt.savefig(data_set+'_azconf.svg')
```

```python
ax = sns.barplot(x="ElConf",
                 y="Frame",
#                  hue="LookName",
                 data=det_list,
                 estimator=lambda x: len(x) / len(det_list)*100)
ax.set(ylabel="Percentage")
plt.title(data_set)
plt.savefig(data_set+'_elconf.svg')
```
